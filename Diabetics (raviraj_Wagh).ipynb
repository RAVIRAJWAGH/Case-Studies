```python
import pandas as pd
import numpy as np
from numpy import mean
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.linear_model import LogisticRegression 
from collections import Counter
from sklearn.svm import SVC
```

```python
dataset=pd.read_csv(r"C:\Users\Parimal\Downloads\diabetes (2).csv")
dataset
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
r1=RandomOverSampler()
x_data,y_data=r1.fit_resample(x,y)
```

```python
print(Counter(y_data))
```

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=5,random_state=11,shuffle=True)
kf.get_n_splits(x_data,y_data)
print(kf)
```

```python
x_data.shape
```

```python
l1=LogisticRegression() # Logistic Regression
```

### KFold

```python
## This loop is only used here to check actualbalancing count of the folds
for train_data,test_data in kf.split(x_data,y_data):

    x_train,x_test=x_data[train_data],x_data[test_data]
    y_train,y_test=y_data[train_data],y_data[test_data]
    
    scores=cross_val_score(l1,x_train,y_train,cv=kf)
    y_pred=cross_val_predict(l1,x_test,y_test)

    print("Checking y_test Balancing.......:",Counter(y_test))
    print(y_pred)
    print(scores)
    print(np.mean(scores)*100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Stratified KFold

```python
from sklearn.model_selection import StratifiedKFold
```

```python
str=StratifiedKFold(n_splits=5,random_state=1,shuffle=True)
```

```python
str.get_n_splits=(x_data,y_data)
```

```python
acc=[]
for train_data,test_data in str.split(x_data,y_data):
    x_train,x_test=x_data[train_data],x_data[test_data]
    y_train,y_test=y_data[train_data],y_data[test_data]
    
    scores=cross_val_score(l1,x_train,y_train,cv=str)
    y_pred=cross_val_predict(l1,x_test,y_test)
    
    print("checking for counter balancing...........................", Counter(y_test))
    print(y_pred)
    print(scores)
    a=mean(scores)*100
    print(a)
    acc.append(a)
print("Accuracies of each fold is :",acc)
```
