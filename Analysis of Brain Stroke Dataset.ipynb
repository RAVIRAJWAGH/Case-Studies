#### Q3) Create new colomn of given categories

```python
abc=[]
for x in data["age"]:
    if x<=15:
        abc.append("kid")
    elif x>15 and x<=30:
        abc.append("teenage")
    elif x>30 and x<=50:
        abc.append("adult")
    else:
        abc.append("old_age")
```

```python
data.insert(2,"age_group",abc)
```
